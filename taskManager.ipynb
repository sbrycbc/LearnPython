```python
# Task Manager

tasks = []  # Görevleri tutmak için bir liste

def add_task():
    """Yeni bir görev ekler."""
    task = input("Lütfen eklemek istediğiniz görevi girin: ")
    tasks.append(task)
    print(f"Görev eklendi: {task}")

def remove_task():
    """Bir görevi listeden kaldırır."""
    view_all_tasks()  # Tüm görevleri göster
    try:
        task_index = int(input("Silmek istediğiniz görevin numarasını girin: ")) - 1
        removed_task = tasks.pop(task_index)
        print(f"Görev silindi: {removed_task}")
    except (IndexError, ValueError):
        print("Geçersiz giriş, lütfen geçerli bir numara girin.")

def view_all_tasks():
    """Tüm görevleri görüntüler."""
    if tasks:
        print("\nTüm Görevler:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")
    else:
        print("\nGörev listesi boş.")

def search_task():
    """Görevlerde arama yapar."""
    keyword = input("Aramak istediğiniz kelimeyi girin: ")
    found_tasks = [task for task in tasks if keyword.lower() in task.lower()]
    if found_tasks:
        print("\nBulunan Görevler:")
        for task in found_tasks:
            print(f"- {task}")
    else:
        print("Hiçbir görev bulunamadı.")

def exit_program():
    """Programdan çıkış yapar."""
    print("Programdan çıkılıyor...")
    return True  # Çıkışı işaret eden True değeri döndür


# Ana program döngüsü
def task_manager():
    exit_flag = False  # Çıkışı işaret etmek için bayrak
    while not exit_flag:  # Döngü, bayrak True olana kadar devam eder
        print("\nTask Manager Menüsü:")
        print("1. Add Task")
        print("2. Remove Task")
        print("3. View All Task")
        print("4. Search Task")
        print("5. Exit")
        choice = input("Seçiminizi yapın (1-5): ")

        if choice == "1":
            add_task()
        elif choice == "2":
            remove_task()
        elif choice == "3":
            view_all_tasks()
        elif choice == "4":
            search_task()
        elif choice == "5":
            exit_flag = exit_program()  # Çıkış yapıldığında bayrak True yapılır
        else:
            print("Geçersiz seçim, lütfen 1-5 arasında bir değer girin.")

# Programı çalıştır
task_manager()

```
